## 텐서 연산

tensor 인덱싱과 슬라이싱

```python
x_tensor = torch.tensor([[1,2,3],[4,5,6]])
x_tensor
```

```python
# 텐서의 인덱싱과 슬라이싱
print('first row:',x_tensor[0])
print('second row:',x_tensor[1])
print('last row:',x_tensor[-1,:])

print('-'*40)

print('first column:',x_tensor[:,0])
print('second column:',x_tensor[:,1])
print('last column:',x_tensor[:,-1])

print('-'*40)

print('second row, last column:', x_tensor[1,2])
print("second row, last column:", x_tensor[1,-1])
```

```python
# 인덱싱해서 데이터 치환하는거 배워보자. 첫번쨰 
print('x_tensor before:',x_tensor)
x_tensor[1,1] = 0

print('-'*40)

print('x_tensor after:',x_tensor)
```

```python
# 인덱싱해서 데이터 치환하는거 배워보자. 두번째 
x_tensor = torch.tensor([[1,2,3],[4,5,6]])

print('x_tensor before:',x_tensor)

print('-'*40)

x_tensor[1] = 0
print('x_tensor after:',x_tensor)
```

tensor concatenate

```python
x_tensor = torch.tensor([[1,2,3],[4,5,6]])

concatenated_dim0 = torch.cat([x_tensor,x_tensor], dim=0)  # concat의 기능은 아래든 옆으로든 갖다 붙이는 것인데, dim=0은 아래로 붙이는 것이다.
print('dim=0:', concatenated_dim0, concatenated_dim0.shape)

print('-'*40)

concatenated_dim1 = torch.cat([x_tensor,x_tensor], dim=1)  # dim=1은 옆으로 붙는 것이다. 
print('dim=1:', concatenated_dim1, concatenated_dim1.shape)
```

tensor 크기 변경

```python
print('x_tensor:', x_tensor, x_tensor.shape)

print('-'*40)

print('x_tensor reshape - 2x3 to 3x2 : ', x_tensor.reshape(3,2), x_tensor.shape)  # 여기서 핵심은 텐서로 만들어진 행렬을 전치 시키는 것이다. 기억 할것은 x_tensor.reshape을 이용하고, 꼭 새로운 변수 지정해야 바뀌는 것을 잊지 마라.
print('x_tensor reshape - 2x3 to 3x2 : ', x_tensor.view(3,2), x_tensor.shape)  # reshape 이나 view는 똑같다. 
```

### Custom 데이터 로더 만들기

```python
from torch.utils.data import Dataset, DataLoader
import torch
import pandas as pd
```

```python
class My_Dataset(Dataset):
    def __init__(self, data, labels): # input으로 data, labels로 받음
        self.data = data
        self.labels = labels
        
    def __len__(self):  # len 데이터 길이 
        return len(self.labels)
        
    def __getitem__(self, idx):  # getitem 이거는 데이터에 인덱스를 매기는 것
        data = self.data[idx]
        label = self.labels[idx]
        sample = {"Data": data, "Class": label}
        return sample
```

```python
# 데이터와 label 정의
data = ['New York', 'Boston', 'California', 'Boston', 'New York']
labels = [1, 2, 3, 2, 1]

# Pandas형태의 DataFrame 정의
Data_frame = pd.DataFrame({'Data': data, 'Labels': labels})

# Dataset class 정의
Data = My_Dataset(Data_frame['Data'],  # 이거는 위에 Class의 My_Dataset에 data, labels 자리에 하나씩 들어간다. 뭐가?  Data_frame['Data'], Data_frame['Labels'] 
                  Data_frame['Labels'])
```

```python
# Display text and label.
print(next(iter(Data)))
print('데이터 수: ', len(data))
print('모든 데이터: ', list(DataLoader(Data)))  # Data안에 데이터를 불러온다음 list로 바꿔서 출력한다. 텐서는 차원을 의미하기도 하네 지금까지 학습한걸로 보면, 
```
